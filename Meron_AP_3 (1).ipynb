### Task 1:  Class Design and Encapsulation

```python
class Student:
    # Constructor
    def init(self, student_id, name, program, gpa):
        self.student_id = student_id
        self.name = name
        self.program = program
        self.__gpa = 0.0
        self.set_gpa(gpa)

    # Getter method
    def get_gpa(self):
        return self.__gpa

    # Setter method
    def set_gpa(self, gpa):
        if 0.0 <= gpa <= 4.0:
            self.__gpa = gpa
        else:
            print("Invalid GPA")

    # Display information
    def display_info(self):
        print("Student ID:", self.student_id)
        print("Name:", self.name)
        print("Program:", self.program)
        print("GPA:", self.__gpa)


# Test Cases   

student1.display_info()

student1.set_gpa(3.9)
print("Updated GPA:", student1.get_gpa())
```

### Task 2: Inheritance with Animal Classes

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print("This animal makes a sound.")
class Dog(Animal):
    def speak(self):
        print(f"{self.name} says: Woof!")
class Cow(Animal):
    def speak(self):
        print(f"{self.name} says: Moo!")
class Bird(Animal):
    def speak(self):
        print(f"{self.name} says: Tweet!")
animals = [
    Dog("The Dog", 3),
    Cow("The Cow", 5),
    Bird("The Bird", 2)
]
for animal in animals:
    animal.speak()
```

#### Report_2
The code defines an Animal class hierarchy that demonstrates inheritance, method overriding, and polymorphism. The parent class Animal contains shared attributes (name and age) and a generic speak  method. Three child classes Dog, Cow, and Bird — inherit from Animal but override the speak method to provide species-specific behavior: dogs say “Woof!”, cows say “Moo!”, and birds say “Tweet!”. In the program, a list of different animal objects is created and stored together, showing how multiple subclasses can be handled uniformly. A loop then iterates through the list and calls speak on each object, and thanks to polymorphism, each animal responds according to its own class definition. This design illustrates how inheritance allows code reuse, while overriding enables customization, and the loop demonstrates polymorphic behavior in practice by treating all animals as instances of the same base type while producing distinct outputs.

### Task 3:  Bioinformatics Sequence System

```python
class Sequence:
    def __init__(self, seq_id, sequence):
        self.seq_id = seq_id
        self.sequence = sequence.upper()
        self.validate()

    def validate(self):
        pass
    def length(self):
        return len(self.sequence)
    def display_info(self):
        print(f"ID: {self.seq_id}")
        print(f"Sequence: {self.sequence}")
        print(f"Length: {self.length()}")
class DNASequence(Sequence):
    def validate(self):
        allowed = set("ATCG")
        if not set(self.sequence).issubset(allowed):
            raise ValueError("Invalid DNA sequence: only A, T, C, G allowed.")
    def gc_content(self):
        gc_count = self.sequence.count("G") + self.sequence.count("C")
        return (gc_count / len(self.sequence)) * 100
class RNASequence(Sequence):
    def validate(self):
        allowed = set("AUCG")
        if not set(self.sequence).issubset(allowed):
            raise ValueError("Invalid RNA sequence: only A, U, C, G allowed.")
    def to_dna(self):
        return self.sequence.replace("U", "T")
class ProteinSequence(Sequence):
    def validate(self):
        allowed = set("ACDEFGHIKLMNPQRSTVWY")  # 20 amino acids
        if not set(self.sequence).issubset(allowed):
            raise ValueError("Invalid protein sequence: only standard amino acids allowed.")
    def count_amino_acids(self):
        counts = {}
        for aa in self.sequence:
            counts[aa] = counts.get(aa, 0) + 1
        return counts
dna = DNASequence("DNA1", "ATCGGCTA")
rna = RNASequence("RNA1", "AUCGGAU")
protein = ProteinSequence("PROT1", "MTEYKLVVVG")
dna.display_info()
print("GC Content:", dna.gc_content(), "%")

rna.display_info()
print("Converted to DNA:", rna.to_dna())

protein.display_info()
print("Amino Acid Counts:", protein.count_amino_acids())
```

#### Report_3
This code defines a parent Sequence class and three child classes — DNASequence, RNASequence, and ProteinSequence — each with its own validation rules and extra methods. When run, it prints the ID, sequence, and length for each type, then shows specialized results the DNA sequence’s GC content, the RNA sequence converted into DNA, and the protein sequence’s amino acid counts.

### Task 4: Library Management System

```python
# Book class
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.is_available = True

    def __str__(self):
        return f"{self.title} by {self.author}"
class Member:
    def __init__(self, member_id, name):
        self.member_id = member_id
        self.name = name
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.is_available:
            book.is_available = False
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed {book.title}.")
        else:
            print(f"Sorry, {book.title} is not available.")

    def return_book(self, book):
        if book in self.borrowed_books:
            book.is_available = True
            self.borrowed_books.remove(book)
            print(f"{self.name} returned {book.title}.")
        else:
            print(f"{self.name} does not have {book.title}.")

    def list_borrowed_books(self):
        if self.borrowed_books:
            print(f"{self.name} has borrowed:")
            for book in self.borrowed_books:
                print(f"- {book.title}")
        else:
            print(f"{self.name} has no borrowed books.")
class Library:
    def __init__(self):
        self.books = []
        self.members = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Added {book.title} to the library.")

    def register_member(self, member):
        self.members.append(member)
        print(f"Registered member: {member.name}")

    def show_books(self):
        print("Library Books:")
        for book in self.books:
            status = "Available" if book.is_available else "Borrowed"
            print(f"- {book.title} ({status})")

library = Library()


book1 = Book("B1", "The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("B2", "Pride and Prejudice", "Jane Austen")
book3 = Book("B3", "The Catcher in the Rye", "J.D. Salinger")
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)


member1 = Member("M1", "Meron Abebe")
member2 = Member("M2", "Sara Tesfaye")
library.register_member(member1)
library.register_member(member2)

member1.borrow_book(book1)
member2.borrow_book(book2)
member1.list_borrowed_books()
member2.list_borrowed_books()

member1.return_book(book1)
library.show_books()
```

#### Report_4
This program is a simple library management system built with Python classes. The Book class stores book details and availability, the Member class manages borrowing and returning books, and the Library class organizes collections of books and members. In the demo, three books are added and two members registered. Each member borrows a book, their borrowed lists are displayed, and one book is returned. Finally, the library shows the updated status of all books. In short, the code demonstrates how object-oriented programming models real-world relationships between books, members, and a library.

### Task 5:  Abstract Base Class and Polymorphism

```python
from abc import ABC, abstractmethod
import math


class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# Circle class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

# Rectangle class
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Triangle class
class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def area(self):
        # Heron's formula
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))

    def perimeter(self):
        return self.a + self.b + self.c
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4, 5)
]
for shape in shapes:
    print(f"{shape.__class__.__name__}:")
    print(f"  Area = {shape.area():.2f}")
    print(f"  Perimeter = {shape.perimeter():.2f}")
    print()
```

#### Report_5
This program defines an abstract Shape class with required methods area() and perimeter(). Child classes (Circle, Rectangle, Triangle) implement these methods according to their geometry. By storing all shapes in a list and looping through them, the program demonstrates abstraction (common interface), polymorphism (different implementations called uniformly), and interface design (forcing consistency across shapes).

### Task 6: Variant Annotation System

```python
# Base class
class Variant:
    def __init__(self, chromosome, position, ref, alt):
        self.chromosome = chromosome
        self.position = position
        self.ref = ref
        self.alt = alt

    def classify(self):
        return "Generic Variant"

    def length(self):
        return abs(len(self.alt) - len(self.ref))

    def summary(self):
        print(f"Chromosome: {self.chromosome}")
        print(f"Position: {self.position}")
        print(f"Reference: {self.ref}")
        print(f"Alternate: {self.alt}")
        print(f"Type: {self.classify()}")
        print(f"Length: {self.length()}")
        print()

# SNP class
class SNP(Variant):
    def classify(self):
        return "SNP"

    def length(self):
        return 1  # SNPs always change one base

# Insertion class
class Insertion(Variant):
    def classify(self):
        return "Insertion"

    def length(self):
        return len(self.alt) - len(self.ref)

# Deletion class
class Deletion(Variant):
    def classify(self):
        return "Deletion"

    def length(self):
        return len(self.ref) - len(self.alt)

# Create variants
variants = [
    SNP("chr1", 12345, "A", "G"),
    Insertion("chr2", 54321, "A", "ATCG"),
    Deletion("chr3", 67890, "ATCG", "A")
]

# Print summaries in one loop
for v in variants:
    v.summary()
```

#### Report_6
It defines a base Variant class with attributes for chromosome, position, reference allele, and alternate allele. It uses inheritance to create specialized classes: SNP, Insertion, and Deletion, each overriding the classify() and length() methods to reflect their biological meaning. By storing all variants in a list and looping through them, the program prints consistent annotation summaries, demonstrating domain modeling, inheritance, and polymorphism in bioinformatics.

### Task 7:  Alternative Constructors

```python
class Genome:
    def __init__(self, name, species, size):
        self.name = name
        self.species = species
        self.size = int(size)  # size in Mb or other unit

    def __str__(self):
        return f"Genome: {self.name}, Species: {self.species}, Size: {self.size}"

    # Class method to create Genome from CSV line
    @classmethod
    def from_csv_line(cls, line):
        parts = line.strip().split(",")
        name, species, size = parts[0], parts[1], parts[2]
        return cls(name, species, size)

    # Class method to create Genome from dictionary
    @classmethod
    def from_dict(cls, data):
        return cls(data["name"], data["species"], data["size"])


# Example usage
csv_input = "NDama,Bos taurus,2800"
dict_input = {"name": "Zebu", "species": "Bos indicus", "size": 3100}

genome1 = Genome.from_csv_line(csv_input)
genome2 = Genome.from_dict(dict_input)

print(genome1)
print(genome2)
```

#### Report_7
It defines a Genome class with a constructor that stores name, species, and size. Two class methods extend its functionality:

from_csv_line parses a comma‑separated string into a Genome object.

from_dict builds a Genome object directly from a dictionary.

By using these methods, the program demonstrates constructors, class methods, and parsing structured data — key skills for bioinformatics workflows where genome metadata often comes in CSV files or JSON dictionaries.

### Task 8: Static Methods vs Class Methods

```python
class DNAUtils:
    # Class variable
    valid_bases = "ATGC"

    # Static method: check if DNA is valid
    @staticmethod
    def is_valid_dna(sequence):
        return all(base in DNAUtils.valid_bases for base in sequence.upper())

    # Static method: compute GC content
    @staticmethod
    def gc_content(sequence):
        sequence = sequence.upper()
        gc_count = sequence.count("G") + sequence.count("C")
        return (gc_count / len(sequence)) * 100 if len(sequence) > 0 else 0

    # Class method: update valid bases
    @classmethod
    def update_valid_bases(cls, new_bases):
        cls.valid_bases = new_bases


# Example usage
seq1 = "ATGCGTA"
seq2 = "ATBX"

print("Seq1:", DNAUtils.is_valid_dna(seq1))
print("Seq1 GC Content:", DNAUtils.gc_content(seq1), "%")

print("Seq2:", DNAUtils.is_valid_dna(seq2))

# Update valid bases (e.g., allow U for RNA)
DNAUtils.update_valid_bases("ATGCU")
print( DNAUtils.is_valid_dna(seq2))
```

#### Report_8
The class variable valid_bases defines which nucleotides are allowed.

The static method is_valid_dna() checks if a sequence contains only valid bases.

The static method gc_content() calculates the percentage of G and C bases.

The class method update_valid_bases() lets you change the allowed bases (e.g., add U for RNA).

This design demonstrates static methods for utility functions, class methods for updating shared state, and class variables for global rules — all essential concepts in Python OOP.

### Task 9: Operator Overloading

```python
# Sequence class with magic methods
class Sequence:
    def __init__(self, seq_id, sequence):
        self.seq_id = seq_id
        self.sequence = sequence.upper()

    # __len__ returns sequence length
    def __len__(self):
        return len(self.sequence)

    # __str__ returns readable info
    def __str__(self):
        return f"Sequence {self.seq_id}: {self.sequence} (Length: {len(self)})"

    # __eq__ compares two sequences
    def __eq__(self, other):
        return self.sequence == other.sequence

    # __add__ joins two sequences
    def __add__(self, other):
        new_seq = self.sequence + other.sequence
        return Sequence(f"{self.seq_id}+{other.seq_id}", new_seq)


# Test cases
seq1 = Sequence("S1", "ATGC")
seq2 = Sequence("S2", "ATGC")
seq3 = Sequence("S3", "GGA")


print("Length of seq1:", len(seq1))
print(seq1)
print("seq1 == seq2?", seq1 == seq2)
print("seq1 == seq3?", seq1 == seq3)
seq4 = seq1 + seq3
print(seq4)
```

#### Report_9
This Sequence class demonstrates Pythonic OOP with magic methods:

len returns the sequence length.

str provides a readable string representation.

eq compares two sequences for equality.

add concatenates two sequences into a new one.

By overriding these methods, the class integrates naturally with Python operators and built‑ins, making sequence objects intuitive to use.
