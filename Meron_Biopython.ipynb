## Exercise 1: Write a GC Content Function

```python
def calculate_gc(sequence):
    """
    Calculate the GC content of a DNA sequence.
    
    Parameters:
        sequence (str): DNA sequence consisting of A, T, G, C.
    
    Returns:
        float: GC content percentage.
    """
    sequence = sequence.upper()  # Ensure consistency
    gc_count = sequence.count("G") + sequence.count("C")
    return (gc_count / len(sequence)) * 100 if len(sequence) > 0 else 0.0

# Example usage:
dna_seq = "ATCGGCTA"
print("GC Content:", calculate_gc(dna_seq), "%")
```

## Exercise 2: Translate Your Own DNA Sequence

```python
dna_seq = "ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG"
# Codon table (simplified)
codon_table = {
    "ATA":"I", "ATC":"I", "ATT":"I", "ATG":"M",
    "ACA":"T", "ACC":"T", "ACG":"T", "ACT":"T",
    "AAC":"N", "AAT":"N", "AAA":"K", "AAG":"K",
    "AGC":"S", "AGT":"S", "AGA":"R", "AGG":"R",
    "CTA":"L", "CTC":"L", "CTG":"L", "CTT":"L",
    "CCA":"P", "CCC":"P", "CCG":"P", "CCT":"P",
    "CAC":"H", "CAT":"H", "CAA":"Q", "CAG":"Q",
    "CGA":"R", "CGC":"R", "CGG":"R", "CGT":"R",
    "GTA":"V", "GTC":"V", "GTG":"V", "GTT":"V",
    "GCA":"A", "GCC":"A", "GCG":"A", "GCT":"A",
    "GAC":"D", "GAT":"D", "GAA":"E", "GAG":"E",
    "GGA":"G", "GGC":"G", "GGG":"G", "GGT":"G",
    "TCA":"S", "TCC":"S", "TCG":"S", "TCT":"S",
    "TTC":"F", "TTT":"F", "TTA":"L", "TTG":"L",
    "TAC":"Y", "TAT":"Y", "TAA":"_", "TAG":"_",
    "TGC":"C", "TGT":"C", "TGA":"_", "TGG":"W",
}

def translate_dna(sequence):
    protein = ""
    for i in range(0, len(sequence), 3):
        codon = sequence[i:i+3]
        if codon in codon_table:
            amino_acid = codon_table[codon]
            if amino_acid == "_":  # Stop codon
                break
            protein += amino_acid
    return protein

# Example usage
print("Translated Protein:", translate_dna(dna_seq))
```

## Exercise 5: Discussion 

 
#### Why is GenBank more informative than FASTA for genome annotation analysis?

FASTA files only store raw sequence data with a simple header, while GenBank files combine the sequence with **rich biological annotations**. This makes GenBank far more useful for genome annotation and functional analysis.

###  Key Differences

| Feature | FASTA | GenBank |
|---------|-------|---------|
| Sequence data | Raw DNA/protein sequence | Full sequence plus annotations |
| Gene names | ❌ Not included | ✅ Each gene is labeled with its name and locus tag |
| CDS coordinates | ❌ Missing | ✅ Start and end positions of coding regions are recorded |
| Protein products | ❌ Not available | ✅ Each CDS includes its translated protein sequence and product name |
| Organism metadata | ❌ None | ✅ Includes species, taxonomy, and strain information |
| References | ❌ None | ✅ Lists publications, authors, and experimental context |

###  Examples

- **FASTA entry**:
geneX
ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG


- **GenBank entry**:
LOCUS       geneX  1200 bp DNA linear
DEFINITION  hypothetical protein geneX
SOURCE      Escherichia coli strain K-12
FEATURES
gene            1..1200
/gene="geneX"
CDS             1..1200
/product="DNA polymerase subunit"
/translation="MAIVMGR..."
REFERENCE   Smith et al., 2024, Journal of Bacteriology



###  Conclusion
- **FASTA** is ideal for quick sequence storage and alignment.  
- **GenBank** is indispensable for **genome annotation, functional analysis, and biological interpretation**, because it includes gene names, CDS coordinates, protein products, organism metadata, and references.


## Exercise 6: Entrez Search Design

```python
from Bio import Entrez

# Always tell NCBI who you are
Entrez.email = "your_email@example.com"  

def entrez_search(query, db="nucleotide", max_results=5):
    """Search NCBI Entrez and return sequence IDs."""
    handle = Entrez.esearch(db=db, term=query, retmax=max_results)
    record = Entrez.read(handle)
    handle.close()
    return record["IdList"]

# Example queries
queries = {
    "Bos taurus mitochondrial genome": '"Bos taurus"[Organism] AND mitochondrion[Filter]',
    "Capra hircus cytochrome b gene": '"Capra hircus"[Organism] AND cytochrome b[Gene]',
    "Ovis aries complete mitochondrial genome": '"Ovis aries"[Organism] AND complete mitochondrial genome',
    "Human BRCA1 gene (research interest)": '"Homo sapiens"[Organism] AND BRCA1[Gene]'
}

# Run searches
for name, query in queries.items():
    ids = entrez_search(query)
    print(f"{name} → {ids}")
```

## Exercise 10 — Improve the Pipeline

```python
from Bio import SeqIO
import pandas as pd

def calculate_gc(seq):
    seq = str(seq).upper()
    gc_count = seq.count("G") + seq.count("C")
    return (gc_count / len(seq)) * 100 if len(seq) > 0 else 0.0

def calculate_at(seq):
    seq = str(seq).upper()
    at_count = seq.count("A") + seq.count("T")
    return (at_count / len(seq)) * 100 if len(seq) > 0 else 0.0

def categorize_gc(gc_content):
    if gc_content >= 60:
        return "High GC"
    elif gc_content >= 40:
        return "Medium GC"
    else:
        return "Low GC"

def analyze_fasta(input_fasta, output_csv):
    results = []
    
    for record in SeqIO.parse(input_fasta, "fasta"):
        seq = record.seq
        seq_string = str(seq).upper()
        
        translated = ""
        stop_codons = 0
        if seq_string.startswith("ATG") and len(seq_string) % 3 == 0:
            translated = str(seq.translate(to_stop=False))
            stop_codons = translated.count("*")
        
        results.append({
            "id": record.id,
            "description": record.description,
            "length": len(seq),
            "gc_content": calculate_gc(seq),
            "at_content": calculate_at(seq),
            "starts_with_ATG": seq_string.startswith("ATG"),
            "length_multiple_of_3": len(seq_string) % 3 == 0,
            "translated_sequence": translated,
            "stop_codons": stop_codons,
            "reverse_complement": str(seq.reverse_complement()),
            "contains_N": "N" in seq_string,
            "gc_category": categorize_gc(calculate_gc(seq))
        })
    
    df = pd.DataFrame(results)
    df.to_csv(output_csv, index=False)
    return df

# Run pipeline
pipeline_results = analyze_fasta("example_sequences.fasta", "sequence_summary.csv")
pipeline_results
```
